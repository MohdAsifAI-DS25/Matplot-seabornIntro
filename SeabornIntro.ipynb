```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.figure(figsize=(10,6))
plt.plot(x,y1,'g-.',label='sin(x)',linewidth=2)
plt.plot(x,y2,'r--',label='cos(x)',linewidth=2)
plt.title('Trigonometric Functions-Matplotlib')
plt.xlabel('sin(x)')
plt.ylabel('cos(x)')
plt.legend()
plt.grid(True,alpha=0.6)
plt.show()
```

#### Now implementing the seaborn

```python
# Import required libraries
import numpy as np          # For numerical operations and creating arrays
import pandas as pd         # For creating and managing dataframes
import seaborn as sns       # For advanced plotting (built on top of matplotlib)
import matplotlib.pyplot as plt  # For creating figures and customizing plots

# ----------------------------
# 1 Generate data
# ----------------------------

# Create 100 evenly spaced values from 0 to 10
x = np.linspace(0, 10, 100)

# Compute sine and cosine values for these x values
y1 = np.sin(x)   # Sine function values
y2 = np.cos(x)   # Cosine function values

# ----------------------------
# 2 Create a DataFrame
# ----------------------------

# Combine the sine and cosine data into a single DataFrame
df = pd.DataFrame({
    'x': np.concatenate([x, x]),               # Repeat x values for both sin and cos
    'y': np.concatenate([y1, y2]),             # Combine sine and cosine y-values
    'function': ['sin(x)']*len(x) + ['cos(x)']*len(x)  # Label each row as sin(x) or cos(x)
})

#  Explanation:
# The DataFrame now looks like this (first few rows):
#       x         y  function
# 0   0.0  0.000000   sin(x)
# 1   0.1  0.099833   sin(x)
# ...
# 99 10.0 -0.544021   sin(x)
# 100 0.0  1.000000   cos(x)
# ...
# 199 10.0 -0.839072  cos(x)

# ----------------------------
# 3 Create the plot
# ----------------------------

plt.figure(figsize=(10, 6))  # Set the figure size (width=10, height=6 inches)

# Use seaborn's lineplot to draw both sine and cosine on the same graph
sns.lineplot(
    data=df,           # The DataFrame to use
    x='x',             # Column for the x-axis
    y='y',             # Column for the y-axis
    hue='function',    # Color lines differently based on the function (sin or cos)
    style='function',  # Different line styles for each function
    markers=True       # Add markers on the data points
)

# ----------------------------
# 4 Add titles and show plot
# ----------------------------

plt.title('Trigonometric Functions - Using Seaborn')  # Set the title of the plot
plt.xlabel('X')                                       # Label for X-axis
plt.ylabel('Y')                                       # Label for Y-axis
plt.grid(True)                                        # Add a grid for better readability
plt.show()                                            # Display the plot
```

### Seaborn Scatter Plot

```python
# Import required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# ----------------------------
# 1 Set random seed for reproducibility
# ----------------------------
np.random.seed(42)  
# This ensures that every time you run the code,
# you get the same random numbers (useful for consistency in results).

# ----------------------------
# 2 Generate random data for x and y
# ----------------------------
x = np.random.randn(100)  
# Creates 100 random numbers from a standard normal distribution (mean=0, std=1)

y = 2 * x + np.random.randn(100) * 0.5  
# Creates a dependent variable y that is roughly a linear function of x (y  2x)
# Adding np.random.randn(100)*0.5 introduces some random noise 
# so the points are scattered around the line rather than being perfectly linear.

# ----------------------------
# 3 Create random categories
# ----------------------------
categories = np.random.choice(['A', 'B', 'C'], size=100)
# Randomly assigns one of three category labels ('A', 'B', or 'C') to each point.
# 'size=100' means we get 100 labels, one for each data point.

# ----------------------------
# 4 Create a pandas DataFrame
# ----------------------------
df_scatter = pd.DataFrame({
    'x': x,                 # x-values
    'y': y,                 # y-values
    'category': categories  # Category labels for color/style
})
# This DataFrame organizes all your data neatly into columns.
# Seaborn will use these column names to map data automatically.

# ----------------------------
# 5 Create the scatter plot
# ----------------------------
plt.figure(figsize=(10, 6))  
# Set the figure size: 10 inches wide, 6 inches tall.

sns.scatterplot(
    data=df_scatter,     # Source DataFrame
    x='x',               # Column for x-axis
    y='y',               # Column for y-axis
    hue='category',      # Color points by category
    style='category',    # Use different marker styles per category
    s=100                # Size of the scatter points
)
# Seaborn automatically handles coloring, legends, and styles based on categories.
# ----------------------------
# 6 Add title and display
# ----------------------------
plt.title("Scatter Plot - Seaborn")  
# Set the title of the plot.
plt.show()  
# Display the scatter plot on screen.
```

## Statistical Plots
### Distribution Plots (Seaborn)


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Generate random data from standard normal distribution (mean=0, std=1)
# Shape: (1000,)  1000 samples
data = np.random.randn(1000)

# Generate 1000 samples from normal distribution: mean=0, std=1
# Equivalent to np.random.randn but more explicit
data1 = np.random.normal(0, 1, 1000)

# Generate 1000 samples from normal distribution: mean=2, std=1.2
# This creates a shifted and slightly wider distribution
data2 = np.random.normal(2, 1.2, 1000)

# Create a large figure with size 15x10 inches for better visibility of subplots
plt.figure(figsize=(15, 10))

# -------------------Subplot 1: Histogram with KDE (Seaborn)---------------------

plt.subplot(2, 3, 1)  # 2 rows, 3 columns, select 1st subplot
sns.histplot(data=data, kde=True, color='skyblue')  # Histogram + Kernel Density Estimate overlay
plt.title('Histogram with KDE - Seaborn')  # Title for the subplot
#  'skyblue' Showing color

# ----------------------------- Subplot 2: KDE Plot only-------------------------
plt.subplot(2, 3, 2)  # Select 2nd subplot
sns.kdeplot(data=data, fill=True, color='lightcoral')  # Smooth density curve, filled underneath
plt.title('KDE Plot')  # Title

# ----------------Subplot 3: Multiple Boxplots (Incorrect Usage) ---------

# Select the 3rd subplot in a 2x3 grid layout
plt.subplot(2, 3, 3)
# Plot two boxplots side by side correctly using a list of datasets
sns.boxplot(data=[data1, data2], palette=['skyblue', 'lightcoral'])
# Add a meaningful title to the plot
plt.title('Comparison of Two Datasets (Boxplots)')
# Add custom x-axis labels for each dataset
plt.xticks([0, 1], ['Dataset 1', 'Dataset 2'])
# Add a y-axis label for clarity
plt.ylabel('Values')
# Show the plot
plt.show()


# --------------------- Subplot 4: Correct Box Plots (Side-by-Side)-------------

plt.subplot(2, 3, 4)  # Select 4th subplot
sns.boxplot(data=[data1, data2])  # Correct: pass list  one box per dataset
plt.title('Box Plot')  # Title
plt.xticks([0, 1], ['Dataset 1', 'Dataset 2'])  # Label x-ticks properly

# -------------------Subplot 5: Violin Plot--------------------------
plt.subplot(2, 3, 5)  # Select 5th subplot
sns.violinplot(data=[data1, data2])  # Shows density shape + boxplot inside
plt.title('Violin Plot')  # Title
plt.xticks([0, 1], ['Dataset 1', 'Dataset 2'])  # Label x-ticks

# --------------------Subplot 6: Swarm Plot----------------------------

plt.subplot(2, 3, 6)  # Select 6th subplot

# Create a tidy DataFrame for swarmplot (long-format)
df_method = pd.DataFrame({
    'value': np.concatenate([data1[:100], data2[:100]]),  # Take only 100 points each
    'dataset': ['Data1'] * 100 + ['Data2'] * 100           # Labels: 100 of each
})

sns.swarmplot(data=df_method, x='dataset', y='value', palette='Set2')  # Individual points, no overlap
plt.title('Swarm Plot')  # Title

# Adjust layout to prevent overlap between subplots
plt.tight_layout()

# Display the entire figure
plt.show()
```

### Categorical Plots


```python
# --------------------------------------------
#                   Step 1: Create category and value lists
# --------------------------------------------

categories = ['A', 'B', 'C', 'D', 'E']   # Define 5 categories (labels)
values = [23, 45, 56, 78, 33]            # Example numeric values (not directly used in plot)

# --------------------------------------------
#        Step 2: Set random seed for reproducibility
# --------------------------------------------

np.random.seed(42)  
# Fixes the random number generator  ensures that every time the code runs, 
# you get the same random data (helps in consistent results and debugging)

# --------------------------------------------
# Step 3: Create a DataFrame using pandas
# --------------------------------------------

df_cat = pd.DataFrame({
    'category': np.repeat(['A', 'B', 'C', 'D'], 50),  
    # np.repeat() repeats each element in the list ['A','B','C','D'] 50 times
    #  total 4  50 = 200 category labels
    
    'value': np.concatenate([
        np.random.normal(10, 2, 50),   # 50 random values for category 'A' (mean=10, std=2)
        np.random.normal(15, 3, 50),   # 50 random values for category 'B' (mean=15, std=3)
        np.random.normal(12, 1, 50),   # 50 random values for category 'C' (mean=12, std=1)
        np.random.normal(18, 2, 50),   # 50 random values for category 'D' (mean=18, std=2)
    ]),
    # np.concatenate() joins these 4 arrays into one continuous array of 200 numeric values
    
    'group': np.tile(['X', 'Y'], 100)   
    # np.tile() repeats the pattern ['X','Y'] 100 times  total 200 group labels
    # So data alternates between X and Y for each observation
})

# --------------------------------------------
# Step 4: Create a figure canvas for multiple subplots
# --------------------------------------------

plt.figure(figsize=(15, 10))
# This initializes a blank figure for plotting
# figsize=(15,10) defines the overall width (15 inches) and height (10 inches)
# This figure will hold multiple subplots (like a grid of small plots)

# ---------------------------- 1.Bar Plot with Error Bars------------------------

# Create the 1st subplot in a grid of 2 rows  3 columns
#  This means the plot will appear in the top-left position of the figure
plt.subplot(2, 3, 1)

# Create a bar plot using Seaborn
sns.barplot(
    data=df_cat,        # The DataFrame that contains the data to be plotted
    x='category',       # Column from df_cat to display on the X-axis (A, B, C, D)
    y='value',          # Column from df_cat to display on the Y-axis (numeric values)
    errorbar='sd',            # Show error bars representing 1 standard deviation (deprecated in new Seaborn)
    capsize=0.1         # Add small horizontal caps at the end of the error bars for clarity
)

# Add a title to this subplot
plt.title('Bar plot with Error bars')


# -------------------------2. Box Plot by Categor ---------------------------

# Create the 2nd subplot in a grid of 2 rows  3 columns
#  This plot will appear in the top-middle position of the figure
plt.subplot(2, 3, 2)

# Create a box plot using Seaborn
sns.boxplot(
    data=df_cat,       # The DataFrame containing your data
    x='category',      # Categorical variable on the X-axis (A, B, C, D)
    y='value',         # Numeric variable on the Y-axis (value distribution)
    hue='group'        # Separate boxes for each group ('X' and 'Y') within each category
)
# Add a title to this subplot
plt.title('Grouped Box Plot')


# ------------------------4. Strip Plot ---------------------------

# Create the 3rd subplot in a 2-row  3-column grid
#  This plot will appear in the bottom-left position of the figure
plt.subplot(2, 3, 3)

# Create a strip plot using Seaborn
sns.stripplot(
    data=df_cat,       # DataFrame containing the data
    x='category',      # Categorical variable on the X-axis (A, B, C, D)
    y='value',         # Numeric variable on the Y-axis (actual data values)
    hue='group',       # Different colors for each group ('X' and 'Y')
    jitter=True        # Adds small random noise along the X-axis to avoid overlapping points
)

# Add a title to this subplot
plt.title('Strip Plot')


# ------------------4. Swarm Plot---------------------------------

# Create the 4th subplot in a 2-row  3-column grid
plt.subplot(2, 3, 4)

# Create a swarm plot using Seaborn
sns.swarmplot(
    data=df_cat,       # The DataFrame that contains the data
    x='category',      # Categorical variable on the X-axis (A, B, C, D)
    y='value',         # Numeric variable on the Y-axis (the actual values)
    hue='group'        # Separate colors for each group ('X' and 'Y')
)

# Add a title to this subplot
plt.title('Swarm Plot')


# --------------5. Point Plot (shows mean values with error----------------------------

# Create the 5th subplot in a 2-row  3-column grid
#  This subplot will appear in the bottom-center of the figure.
plt.subplot(2, 3, 5)

# Create a point plot using Seaborn
sns.pointplot(
    data=df_cat,        # DataFrame containing the data
    x='category',       # Categorical variable on the X-axis (A, B, C, D)
    y='value',          # Numeric variable on the Y-axis
    hue='group',        # Separate lines/points for each group ('X' and 'Y')
    errorbar='sd',      # Show standard deviation as error bars (replaces deprecated ci='sd')
    capsize=0.1         # Add small horizontal caps at the ends of error bars
)

# Add a title to this subplot
plt.title("Point Plot")

# Adjust spacing between subplots to prevent overlap
plt.tight_layout()

# Display all the plots in the current figure window
plt.show()
```

# Advanced Statistical Plots
### Regression Plots


```python
# -------------------------------------------------------
#                     Generate sample regression data
# -------------------------------------------------------
np.random.seed(42)  # Set random seed for reproducibility (so results are the same each run)
x = np.random.randn(100)  # 100 random values from a standard normal distribution
y = 2*x + 1 + np.random.randn(100)*0.5  # Linear relationship (y = 2x + 1 + some noise)

# Create a DataFrame and randomly assign each row to group 'A' or 'B'
df_reg = pd.DataFrame({
    'x': x,
    'y': y,
    'group': np.random.choice(['A', 'B'], 100)
})

# -------------------------------------------------------
#                  Create a figure with 3 subplots (side-by-side)
# -------------------------------------------------------
plt.figure(figsize=(15,5))  # Define overall figure size (width=15, height=5 inches)

# -------------------------------------------------------
#                       (1) Simple regression plot
# -------------------------------------------------------
plt.subplot(1,3,1)  # 1 row, 3 columns, first subplot
sns.regplot(data=df_reg, x='x', y='y', scatter_kws={'alpha':0.6})  # Scatter + regression line
plt.title("Simple Regression Plot")  # Title for subplot

# -------------------------------------------------------
#                        (2) Grouped regression plot (manual method)
# -------------------------------------------------------
plt.subplot(1,3,2)  # 1 row, 3 columns, second subplot

# Loop through each group and plot its regression line separately
for g, d in df_reg.groupby('group'):
    sns.regplot(
        data=d, x='x', y='y', label=g, scatter_kws={'alpha':0.6}
    )

plt.legend()  # Show legend to distinguish groups A and B
plt.title("Grouped Regression Plot")  # Title for subplot

# -------------------------------------------------------
#                        (3) Residual plot
# -------------------------------------------------------
plt.subplot(1,3,3)  # 1 row, 3 columns, third subplot
sns.residplot(data=df_reg, x='x', y='y', scatter_kws={'alpha':0.6})  # Plot residuals
plt.title("Residual Plot")  # Title for subplot

# -------------------------------------------------------
#                    Adjust layout and display all subplots
# -------------------------------------------------------
plt.tight_layout()  # Adjust spacing between subplots to avoid overlap
plt.show()          # Render the complete figure
```

#### Seaborn Multi-Plot Grids.......


```python
# Create a DataFrame with random values
df_grid = pd.DataFrame({
    'x': np.random.randn(200),                         # Random values for X-axis
    'y': np.random.randn(200),                         # Random values for Y-axis
    'category': np.random.choice(['A', 'B', 'C'], 200),# Randomly assign category A, B, or C
    'values': np.random.exponential(2, 200)            # Random exponential values (not used in this plot)
})

# -------------------------------------------------------
# FacetGrid: multiple scatter plots divided by 'category'
# -------------------------------------------------------
# Create a FacetGrid object with one subplot per category
g = sns.FacetGrid(df_grid, col='category', col_wrap=2, height=4)

# Map a scatter plot to each facet (one per category)
g.map_dataframe(sns.scatterplot, x='x', y='y', alpha=0.7)

# Add subplot titles showing the category name
g.set_titles('{col_name}')

# Add a legend (useful if we color points differently)
g.add_legend()

# Display the FacetGrid plot
plt.show()

# -------------------------------------------------------
# PairPlot: visualize pairwise relationships across features
# -------------------------------------------------------
# Load the classic 'iris' dataset (built into Seaborn)
iris = sns.load_dataset('iris')

# Create a pair plot to show relationships between numeric features
sns.pairplot(
    iris,
    hue='species',                # Color points by species
    diag_kind='hist',             # Use histogram on the diagonal
    markers=['o', 's', 'D']       # Different markers for each species
)

# Add a main title slightly above the figure
plt.suptitle('Pair Plot - Seaborn', y=1.02)

# Show the pair plot
plt.show()

# -------------------------------------------------------
# JointPlot: relationship between two variables with marginal histograms
# -------------------------------------------------------
sns.jointplot(
    data=iris,
    x='sepal_length',
    y='sepal_width',
    hue='species',      # Color points by species
    kind='scatter',     # Scatter plot in the center
    alpha=0.7           # Slight transparency
)

# Add a main title above the joint plot
plt.suptitle('Joint Plot - Seaborn', y=1.02)

# Display the joint plot
plt.show()
```

#### Styling and Custmization.....
###### Seabron Styling

```python
# -------------------------------------------------------
# Seaborn Styling Options Demo
# -------------------------------------------------------

# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Define lists of Seaborn built-in style themes and color palettes
# 'style' changes the background grid and axes appearance
# 'palette' changes the color scheme used in plots
styles = ['whitegrid', 'darkgrid', 'white', 'dark', 'ticks']
palettes = ['deep', 'muted', 'pastel', 'bright', 'dark']  # (fixed typo: colorblind  dark for matching length)

# Create a large figure with 2 rows and 3 columns of subplots
plt.figure(figsize=(15, 10))

# Loop through each stylepalette combination and make a small bar plot
for i, (style, palette) in enumerate(zip(styles, palettes)):
    # 1 Apply the chosen style and color palette
    sns.set_theme(style=style, palette=palette)

    # 2 Select the position of the current subplot
    plt.subplot(2, 3, i + 1)

    # 3 Create sample categorical data
    x = ['A', 'B', 'C', 'D']
    y = [10, 15, 12, 18]

    # 4 Plot a simple bar chart using Seaborn
    sns.barplot(x=x, y=y)

    # 5 Add a title showing which style and palette are applied
    plt.title(f"style: {style}\npalette: {palette}")

# 6 Adjust spacing so subplots dont overlap
plt.tight_layout()

# 7 Display the full grid of styled bar charts
plt.show()

# -------------------------------------------------------
# Reset theme after demo
# -------------------------------------------------------

# Set a default theme for future plots (whitegrid is clean and common)
sns.set_theme(style='whitegrid')
```


## Real Dataset Examples
##### Seaborn 

```python
tips=sns.load_dataset('tips')
print(tips.head())
plt.figure(figsize=(15,12))

#Matplotlib version
plt.subplot(2,3,1)
days=tips['day'].unique()
gender_totals=tips.groupby(['day','sex'],observed=False)['total_bill'].mean().unstack()
width=0.35
x=np.arange(len(days))

plt.bar(x-width/2,gender_totals['Male'],width,label='Male',alpha=0.7)
plt.bar(x+width/2,gender_totals['Female'],width,label='Female',alpha=0.7)
plt.xlabel('Day')
plt.ylabel('Average Total Bill')
plt.title('Matplotlib:Average Bill By day & Gender')
plt.xticks(x,days)
plt.legend()

#Seaborn versions
plt.subplot(2,3,2)
sns.barplot(data=tips,x='day',y='total_bill',hue='sex',errorbar='sd',capsize=0.1)
plt.title('Seaborn:Bar Plot')

plt.subplot(2,3,3)
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
plt.title('Seaborn:Box plots')

plt.subplot(2,3,4)
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True)
plt.title('Seaborn:Violin plots')

plt.subplot(2,3,5)
sns.scatterplot(data=tips,x='day',y='total_bill',hue='day',style='sex',s=100)
plt.title('Seaborn:Scatter plots')

plt.subplot(2,3,6)
sns.heatmap(tips.select_dtypes(include='number'),annot=True,cmap='coolwarm',center=0)
plt.title('Seaborn:correlation Heatmap')

plt.tight_layout()
plt.show()
```
