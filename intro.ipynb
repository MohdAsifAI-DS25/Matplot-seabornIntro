# Line plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
```

```python
# Create a new figure with a specific size (width=10, height=2 inches)
plt.figure(figsize=(10, 2))

# Plot the data: x vs y, line width 2, labeled 'sin(x)', and color red
plt.plot(x, y, linewidth=2, label='sin(x)', color='red')

# Add a title to the plot
plt.title('Sine wave')

# Label the X-axis
plt.xlabel('X axis')

# Label the Y-axis
plt.ylabel("Y axis")

# Display the legend (shows the label 'sin(x)')
plt.legend()

# Enable the grid for better readability
plt.grid(True)

# Display the plot on the screen
plt.show()
```

# Scatter Plot

```python
x = np.random.randn(100)      # 100 random x-values (normal distribution)
y = np.random.randn(100)      # 100 random y-values (normal distribution)
colors = np.random.randn(100) # 100 random color values (used in scatter plot)
sizes = 1000 * np.random.randn(100) # 100 random sizes (scaled by 1000)
```

### Show the data

```python
# Create a new figure with a specific size (8 inches wide, 6 inches tall)
plt.figure(figsize=(8,6))

# Create a scatter plot
# x, y  → coordinates of each point
# c=colors  → color values (mapped using the colormap)
# s=sizes   → size of each point
# alpha=0.6 → transparency level (0 = invisible, 1 = solid)
# cmap='viridis' → colormap used for coloring the points
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')

# Add a colorbar (legend for the colors), labeling it 'Color Intensity'
plt.colorbar(label='Color Intensity')

# Add a title to the plot
plt.title("Scatter Plot with Color and Size")

plt.grid(True)

# Label the X-axis
plt.xlabel('X Values')

# Label the Y-axis
plt.ylabel('Y Values')

# Display the final plot on the screen
plt.show()
```

# Mutliple Plots and Subplots
### Mutiple Lines on Same plots

```python
# Create an array of 100 evenly spaced values between 0 and 10
x = np.linspace(0, 10, 100)

# Calculate the sine of each x value
y1 = np.sin(x)

# Calculate the cosine of each x value
y2 = np.cos(x)

# Calculate the tangent of each x value
y3 = np.tan(x)
```

```python
# Create a new figure with a specific size (10 inches wide, 6 inches tall)
plt.figure(figsize=(10,6))

# Plot the sine function: x vs y1, red solid line ('r-'), with label 'sin(x)'
plt.plot(x, y1, 'r-', label='sin(x)')

# Plot the cosine function: x vs y2, blue solid line ('b-'), with label 'cos(x)'
plt.plot(x, y2, 'b--', label='cos(x)')

# Plot the tangent function: x vs y3, green solid line ('g-'), with label 'tan(x)'
plt.plot(x, y3, 'g:', label='tan(x)')

# Add a title to the plot
plt.title("Trigonometric Functions")

# Label the X-axis
plt.xlabel('X axis')

# Label the Y-axis
plt.ylabel("Y axis")

# Display the legend to identify each curve (sin, cos, tan)
plt.legend()

# Limit the Y-axis between -2 and 2 (so the tan spikes don’t distort the plot)
plt.ylim(-2, 2)

# Display a grid for better readability
plt.grid(True)

# Show the final plot on the screen
plt.show()
```

#### Subplots

```python
# Create a 2x2 grid of subplots (4 plots total) inside one figure
# The figure size is 12 inches wide and 8 inches tall
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# -------------------------------------------------------------------
# PLOT 1: Line Plot (Top-Left)
# -------------------------------------------------------------------

# Create 100 evenly spaced x values from 0 to 10
x = np.linspace(0, 10, 100)

# Plot a sine wave in the first subplot (top-left)
# 'r-' means a red solid line
axes[0, 0].plot(x, np.sin(x), 'r-')

# Set the title for the first subplot
axes[0, 0].set_title('Sine Wave')

# Label the X-axis and Y-axis
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('sin(x)')

# -------------------------------------------------------------------
# PLOT 2: Scatter Plot (Top-Right)
# -------------------------------------------------------------------

# Generate 50 random x and y values from a normal distribution
x = np.random.randn(50)
y = np.random.randn(50)

# Create a scatter plot with some transparency (alpha=0.6)
axes[0, 1].scatter(x, y, alpha=0.6)

# Set the title and axis labels for the scatter plot
axes[0, 1].set_title('Scatter Plot')
axes[0, 1].set_xlabel('X')
axes[0, 1].set_ylabel('Y')

# -------------------------------------------------------------------
# PLOT 3: Bar Chart (Bottom-Left)
# -------------------------------------------------------------------

# Define the category labels and their corresponding values
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 85]

# Create a bar chart with custom colors for each bar
axes[1, 0].bar(categories, values, color=['red', 'blue', 'green', 'orange', 'violet'])

# Set the title and axis labels for the bar chart
axes[1, 0].set_title('Bar Chart')
axes[1, 0].set_xlabel("Categories")
axes[1, 0].set_ylabel("Values")

# -------------------------------------------------------------------
# PLOT 4: Histogram (Bottom-Right)
# -------------------------------------------------------------------

# Generate 1000 random numbers from a normal distribution
data = np.random.randn(1000)

# Create a histogram with 30 bins, semi-transparent bars (alpha=0.6),
# and black edges around each bar
axes[1, 1].hist(data, bins=30, alpha=0.6, edgecolor='black')

# Set the title and axis labels for the histogram
axes[1, 1].set_title('Histogram')
axes[1, 1].set_xlabel("Values")
axes[1, 1].set_ylabel("Frequency")

# -------------------------------------------------------------------
# Adjust the layout to prevent overlapping titles and labels
plt.tight_layout()

# Display all four subplots on the screen
plt.show()
```

# Different Plot Types
### Bar Charts

```python
# Define category names for the X-axis
categories = ['categories A', 'categories B', 'categories C', 'categories D']
# Define values for two series (two bars per category)
values1 = [23, 45, 55, 78]
values2 = [34, 38, 62, 45]
# Create an array of X positions (0, 1, 2, 3)
x = np.arange(len(categories))
# Set the width of each bar
width = 0.5
# Create a new figure with a specific size (10x6 inches)
plt.figure(figsize=(10,6))
# Plot the first set of bars (values1) slightly to the left of each category position
plt.bar(x - width/2, values1, width, label='Series 1', alpha=0.6)
# Plot the second set of bars (values2) slightly to the right of each category position
plt.bar(x + width/2, values2, width, label='Series 2', alpha=0.8)
# Label the X and Y axes
plt.xlabel("Categories")
plt.ylabel("Values")
# Add a title for the entire chart
plt.title('Grouped Bar Chart')
# Replace numeric X positions with category names
plt.xticks(x, categories)
# Show the legend to differentiate Series 1 and Series 2
plt.legend()
# Display the chart
plt.show()


#HORIZONTAL BAR GRAPH

# Create a new figure for the horizontal bar chart
plt.figure(figsize=(10,6))
# Create horizontal bars using barh()
plt.barh(categories, values1, alpha=0.7)
# Label the X-axis
plt.xlabel('Values')
# Add a title for the chart
plt.title("Horizontal Bar Chart")
# Display the chart
plt.show()

```

# Pie Chart

```python
# Define the labels (names) for each slice of the pie chart
labels = ['Python', 'Java', 'C++', 'SQL', 'JavaScript', 'Other']

# Define the corresponding sizes (numerical values) for each slice
sizes = [35, 15, 20, 15, 5, 10]
# Generate 6 random RGB color values (each row is one color)
# np.random.rand(6,3) → creates an array of 6 colors, each with 3 RGB values between 0 and 1
colors = np.random.rand(6, 3)
# Define how much each slice should be "exploded" (separated from the center)
# All zeros mean no slice is pulled out
explode = (0, 0, 0, 0, 0, 0)
# Create a new figure (canvas) with a size of 8x8 inches
plt.figure(figsize=(8, 8))
# Create the pie chart with the given data and formatting options
plt.pie(
    sizes,                # The data (numerical values for each slice)
    explode=explode,      # How much to offset each slice
    labels=labels,        # Names of the slices
    colors=colors,        # Randomly generated colors for each slice
    autopct='%1.1f%%',    # Display percentages on each slice (1 decimal place)
    shadow=True,          # Add a shadow for a 3D look
    startangle=90         # Start drawing the pie chart at a 90° angle (top)
)

# Make sure the pie is drawn as a perfect circle (not an oval)
plt.axis('equal')

# Add a title to the chart
plt.title('Programming Language Popularity')

# Display the pie chart on screen
plt.show()
```

## Box Plots


```python
# Generate three datasets using random normal distributions
# Each dataset has 100 values, with increasing standard deviation (1, 2, and 3)
# np.random.normal(mean, std, size)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
# Create a new figure (canvas) with a size of 8x6 inches
plt.figure(figsize=(8, 6))
# Create the box plots for the three datasets
# 'labels' assigns names for each dataset shown below each box
plt.boxplot(data, labels=['Dataset 1', 'Dataset 2', 'Dataset 3'])
# Add a title to the box plot
plt.title("Box Plots Example")
# Label the Y-axis to indicate the values
plt.ylabel('Values')
# Add a grid to make it easier to read values across the chart
# alpha=0.6 makes the grid slightly transparent
plt.grid(True, alpha=0.6)
# Display the final box plot on the screen
plt.show()
```

# Customizaton and Styling
## colors,Markers and Line Styles

```python
#Different Line Styles
# ------------------- (1) DIFFERENT LINE STYLES -------------------
# Create 20 evenly spaced values between 0 and 10 for the x-axis
x = np.linspace(0, 10, 20)

# Create a new figure (canvas) with a specific size (12 inches wide, 8 inches tall)
plt.figure(figsize=(12, 8))

# Create a subplot in position (2 rows, 2 columns, 1st plot = top-left corner)
plt.subplot(2, 2, 1)

# Plot different sine curves using various line styles:
plt.plot(x, np.sin(x), 'r-', label='solid')     # Red solid line ('r-')
plt.plot(x, np.sin(x + 1), 'b--', label='dashed') # Blue dashed line ('b--')
plt.plot(x, np.sin(x + 2), 'g:', label='dotted')  # Green dotted line ('g:')
plt.plot(x, np.sin(x + 3), 'y-.', label='dashdot') # Yellow dash-dot line ('y-.')

# Add a title for this subplot
plt.title('Line Styles')

# Display a legend to label each line style
plt.legend()


# ------------------- (2) DIFFERENT MARKERS -------------------

# Create the second subplot (top-right position in a 2x2 grid)
plt.subplot(2, 2, 2)
# Plot a sine curve with red circle markers ('ro')
# 'r' = red color, 'o' = circle marker
plt.plot(x, np.sin(x), 'ro', label='circle')
# Plot another sine curve shifted by +1 on x, using blue triangle markers ('b^')
# 'b' = blue color, '^' = triangle marker
plt.plot(x, np.sin(x + 1), 'b^', label='triangle')
# Plot a third sine curve shifted by +2 on x, using green square markers ('gs')
# 'g' = green color, 's' = square marker
plt.plot(x, np.sin(x + 2), 'gs', label='square')
# Plot a fourth sine curve shifted by +3 on x, using magenta star markers ('m*')
# 'm' = magenta color, '*' = star marker
plt.plot(x, np.sin(x + 3), 'm*', label='star')
plt.grid(True)
# Add a title to this subplot
plt.title('Marker Style')
# Display a legend to identify each marker type
plt.legend()


# ------------------- (3) COLOR MAP (HEATMAP) -------------------

# Create the third subplot (bottom-left position in a 2x2 grid)
plt.subplot(2, 2, 3)

# Generate a 10x10 matrix of random numbers between 0 and 1
# This data will be visualized as colors in a heatmap
y = np.random.rand(10, 10)

# Display the 2D data (matrix) as an image
# 'cmap' defines the color scheme — here we use the 'hot' color map (red–yellow–black tones)
plt.imshow(y, cmap='magma')

# Add a color bar beside the plot to indicate the data-to-color mapping
plt.colorbar()

# Add a title for this subplot
plt.title('Heatmap with Colorbar')


# ------------------- (4) TEXT AND ANNOTATIONS -------------------

# Create the fourth subplot (bottom-right position in a 2x2 grid)
plt.subplot(2, 2, 4)

# Generate X values evenly spaced between 0 and 10 (100 points)
x = np.linspace(0, 10, 100)

# Calculate the sine of each X value
y = np.sin(x)

# Plot the sine wave using a blue solid line ('b-')
plt.plot(x, y, 'b-.')

# Add title and axis labels
plt.title('Text and Annotations')
plt.xlabel('X axis')
plt.ylabel('Y axis')

# Add a text label directly on the plot at coordinates (x=2, y=0.5)
plt.text(2, 0.5, 'Important Point', fontsize=12, color='red')

# Add an annotation (arrow + label) to highlight a specific point on the curve
# xy = point being annotated (x=π/2, y=1)
# xytext = position of the annotation text (x=π/2+1, y=0.8)
# arrowprops = dictionary defining arrow color and shape
plt.annotate(
    'Maximum', 
    xy=(np.pi/2, 1), 
    xytext=(np.pi/2 + 1, 0.8),
    arrowprops=dict(facecolor='black', shrink=0.075)
)
# Adjust the spacing between subplots to prevent overlap of titles and labels
plt.tight_layout()
# Display all subplots together
plt.show()



```

# Advanced Features
### 3D plots

```python
# Import the necessary library for 3D plotting
from mpl_toolkits.mplot3d import Axes3D
# Create a figure with specific size (width=10, height=8)
fig = plt.figure(figsize=(10, 8))

# Add a 3D subplot to the figure (1 row, 1 column, position 1)
ax = fig.add_subplot(111, projection='3d')

# Generate 1D arrays of X and Y values between -5 and 5 (100 points each)
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create 2D grid coordinates from the 1D arrays
# meshgrid creates matrices X and Y for surface plotting
X, Y = np.meshgrid(x, y)

# Calculate Z values based on X and Y
# sqrt(X^2 + Y^2) makes circular wave pattern
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D surface plot with a color map ('hot') and 90% opacity
surface = ax.plot_surface(X, Y, Z, cmap='magma', alpha=0.9)

# Label the X, Y, and Z axes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Set a title for the plot
ax.set_title('3D Surface Plot')

# Add a color bar to show the value-to-color relationship
# shrink=0.5 reduces size, aspect=5 changes its shape
fig.colorbar(surface, shrink=0.5, aspect=5)

# Display the final 3D plot
plt.show()
```

### Polar plots

```python
theta=np.linspace(0,2*np.pi,100)
r=2+np.sin(5*theta)

plt.figure(figsize=(8,8))
ax=plt.subplot(111,projection='polar')
ax.plot(theta,r,'r-.',linewidth=2)
ax.set_title('Polar Plot Example',va='bottom')
plt.show()
```
